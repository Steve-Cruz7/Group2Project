```python
import numpy as np # We need np to load the npz datasets
```

```python
real_data = np.load('data/real_world/malware.npz')
# Real World Data vanilla testing
```

```python
# Grabbing Data
train_data = real_data['X_train']
train_labels = real_data['y_train']
test_data = real_data['X_test']
test_labels = real_data['y_test']
```

```python
# Reading some of the data
print(train_data)
```

```python
# Very vanilla MLP
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes = (512, 256)) # activation function is relu and optimizer is adam, default loss is cross entropy
```

```python
mlp.fit(train_data, train_labels)
```

```python
predictions = mlp.predict(test_data)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_labels, predictions)
```
